```python
# Counting Ratings
plt.figure(figsize = (10,6))
rating = df['IMDB_Rating'].value_counts().index
rating_count = df['IMDB_Rating'].value_counts().values
plt.xlabel('IMDB Rating')
plt.ylabel('No of Movies')
plt.title('Ratings Count')
sns.pointplot(x = rating, y = rating_count)
plt.grid()
```

The line chart shows that when the number of movies are in between 120-140, the IMDB rating is 7.6. When the number of movies are closed to 160,the IMDB rating is 7.7. When the number of movies are in between 140-160,the IMDB rating is 7.8. When the number of movies are in between 100-120,the IMDB rating is 7.9. When the number of movies are 140,the IMDB rating is 8.0. After the number of movies decreased bt the IMDB rating is increased upto 9.3. That means a small amount of movies got highest IMDB rating of 9.3.

```python
# Number of movies produced per year
plt.figure(figsize = (15,10))
year = df['Released_Year'].value_counts().sort_values().index
year_count = df['Released_Year'].value_counts().sort_values().values
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies per Year')
sns.pointplot(x = year, y = year_count)
plt.xticks(rotation = 90)
plt.grid()
```

```python
# AVERAGE IMDB RATING PER DIRECTOR
average_rating_per_director = df.groupby('Director')['IMDB_Rating'].mean()

# Sort directors based on average IMDb rating in descending order
sorted_directors_by_rating = average_rating_per_director.sort_values(ascending=False)

# Select the top 10 directors
top_10_directors_by_rating = sorted_directors_by_rating.head(10)

# Plotting
plt.figure(figsize=(10, 6))
top_10_directors_by_rating.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Directors by Average IMDB Rating')
plt.xlabel('Director')
plt.ylabel('Average IMDB Rating')
plt.show()
```

in the histogram bar graph the director Frank Darabont whose average IMDB rating is more than 8 the second highest there are two director Irvin Kershnner and Femando Meirelles.
