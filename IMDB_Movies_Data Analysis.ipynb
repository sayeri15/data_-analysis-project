```python
import pandas as pd       # Importing Pandas as pd
import numpy as np        # Importing Numpy as np
import seaborn as sns     # Importing Seaborn as sns
import matplotlib.pyplot as plt      # Importing Pyplot sub-module of Matplotlib library as plt
```

```python
df=pd.read_csv('imdb_top_1000.csv')    # Reading csv file
```

```python
dup_data=df.duplicated()
print("Duplicate values:",dup_data)
df=df.drop_duplicates()
```

```python
df.columns
```

```python
df.groupby('Released_Year')['No_of_Votes'].mean().sort_values(ascending=False)  
```

```python
sns.barplot(x='Released_Year',y='No_of_Votes',data=df)
plt.title("Votes By Year")
plt.show()
```

With respect to Released Year, the No of votes are fractuating.After a certain time it decreases but suddenly it increases.

```python
df['Released_Year'].value_counts()
```

```python
sns.countplot(x='Released_Year',data=df)
plt.title('Number of Movies per Year')
plt.show()
```

Movies count is fractuating with Released year. But the graph shows the count is decreased at the end.

```python
top10_len=df.nlargest(10,'IMDB_Rating')[['Series_Title','IMDB_Rating','Director']]\
.set_index('Series_Title')
```

```python
top10_len
```

Here it shows the top 10 movie titles with IMDB rating and the names of the directors.

```python
sns.barplot(x='IMDB_Rating',y=top10_len.index,data=top10_len,hue='Director')
plt.show()
```

```python
df.groupby('Released_Year')['IMDB_Rating'].mean().sort_values(ascending=False)
```

```python
# Counting Ratings
plt.figure(figsize = (10,6))
rating = df['IMDB_Rating'].value_counts().index
rating_count = df['IMDB_Rating'].value_counts().values
plt.xlabel('IMDB Rating')
plt.ylabel('No of Movies')
plt.title('Ratings Count')
sns.pointplot(x = rating, y = rating_count)
plt.grid()
```

The line chart shows that at 7.6 IMDB rating, the number of movies is above 120. In 7.7 IMDB rating,number of movies is the highest,almost 160. At 7.8 rating, number of movies is between 140-160. At 7.9 rating, it is between 100-120. At 8.0 rating, it is 140. After 8.0, number of movies are dropped and at 9.3 rating it is the lowest.

```python
# Number of movies produced per year
plt.figure(figsize = (15,10))
year = df['Released_Year'].value_counts().sort_values().index
year_count = df['Released_Year'].value_counts().sort_values().values
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies per Year')
sns.pointplot(x = year, y = year_count)
plt.xticks(rotation = 90)
plt.grid()
```

```python
df['Gross'].dtype
```

```python
# Handling Null Variables
df['Gross'] = df['Gross'].str.replace(',','')
df['Gross'] = df['Gross'].replace(np.nan, 0)
df['Gross'].head()
```

```python
df[['IMDB_Rating','Meta_score','No_of_Votes','Gross']].corr()
```

```python
correlations = df[['IMDB_Rating','Meta_score','No_of_Votes','Gross']].corr()
sns.heatmap(data = correlations, cmap = 'coolwarm', annot = True)
```

```python
# Top Voted Movies
top_voted = df.sort_values(['No_of_Votes'], ascending = False)
top_voted.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x = top_voted['Series_Title'][:10], y = top_voted['No_of_Votes'][:10])
plt.xticks(rotation = 90)
plt.tight_layout()
plt.xlabel('Movie Title')
plt.ylabel('Number of Votes')
plt.title('Top Voted Movies')
plt.show()
```

```python
# Top Voted Ratings
plt.figure(figsize=(10,6))
sns.barplot(x = top_voted['Series_Title'][:10], y = top_voted['IMDB_Rating'][:10])
plt.xticks(rotation = 90)
plt.tight_layout()
plt.xlabel('Movie Title')
plt.ylabel('IMDB Rating')
plt.title('IMDB Rating of Top Voted Movies')
plt.show()
```
