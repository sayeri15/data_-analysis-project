```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_username', 'your_password', 'your_host', and 'your_database' with actual credentials
connection = mysql.connector.connect(
    user='root',
    password='ROOT',
    host='localhost',
    database='ecommerce'
)

# Create a cursor object to execute SQL queries
cursor = connection.cursor()

# Query to identify repeat customers and their order patterns
repeat_customers_query = '''
SELECT c.customer_id, c.name, COUNT(DISTINCT od.order_id) as order_count
FROM customer c
JOIN order_details od ON c.customer_id = od.customer_id
GROUP BY c.customer_id, c.name
HAVING order_count > 1
'''

# Execute the query
cursor.execute(repeat_customers_query)

# Fetch the result into a Pandas DataFrame
repeat_customers_data = pd.DataFrame(cursor.fetchall(), columns=['Customer ID', 'Customer Name', 'Order Count'])

# Plotting the order patterns of repeat customers using Matplotlib
plt.figure(figsize=(12, 6))
plt.bar(repeat_customers_data['Customer Name'], repeat_customers_data['Order Count'], color='blue')
plt.title('Order Patterns of Repeat Customers')
plt.xlabel('Customer Name')
plt.ylabel('Order Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Query to calculate customer retention rates over time
customer_retention_rate_query = '''
SELECT YEAR(order_date) as order_year, MONTH(order_date) as order_month,
       COUNT(DISTINCT customer_id) as total_customers,
       COUNT(DISTINCT CASE WHEN YEAR(NOW()) - YEAR(order_date) = 0 AND MONTH(NOW()) - MONTH(order_date) <= 1 THEN customer_id END) as retained_customers
FROM order_details
GROUP BY order_year, order_month
ORDER BY order_year, order_month
'''

# Execute the query
cursor.execute(customer_retention_rate_query)

# Fetch the result into a Pandas DataFrame
customer_retention_rate_data = pd.DataFrame(cursor.fetchall(), columns=['Order Year', 'Order Month', 'Total Customers', 'Retained Customers'])

# Calculate customer retention rate
customer_retention_rate_data['Retention Rate (%)'] = (customer_retention_rate_data['Retained Customers'] / customer_retention_rate_data['Total Customers']) * 100

# Plotting the customer retention rates over time using Matplotlib
plt.figure(figsize=(12, 6))
plt.plot(customer_retention_rate_data['Order Year'].astype(str) + '-' + customer_retention_rate_data['Order Month'].astype(str),
         customer_retention_rate_data['Retention Rate (%)'], marker='o', color='orange', linestyle='-', linewidth=2)
plt.title('Customer Retention Rates Over Time')
plt.xlabel('Order Year-Month')
plt.ylabel('Retention Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Close the cursor and connection
cursor.close()
connection.close()
```

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_username', 'your_password', 'your_host', and 'your_database' with actual credentials
connection = mysql.connector.connect(
    user='root',
    password='ROOT',
    host='localhost',
    database='ecommerce'
)

# Create a cursor object to execute SQL queries
cursor = connection.cursor()

# Query to analyze the performance of different products in terms of sales
product_sales_analysis_query = '''
SELECT od.product_id, p.product_name, SUM(od.quantity) as total_sales
FROM order_details od
JOIN product p ON od.product_id = p.product_id
GROUP BY od.product_id, p.product_name
'''

# Execute the query
cursor.execute(product_sales_analysis_query)

# Fetch the result into a Pandas DataFrame
product_sales_analysis_data = pd.DataFrame(cursor.fetchall(), columns=['Product ID', 'Product Name', 'Total Sales'])

# Plotting the performance of different products in terms of sales using Matplotlib
plt.figure(figsize=(12, 6))
plt.bar(product_sales_analysis_data['Product Name'], product_sales_analysis_data['Total Sales'], color='skyblue')
plt.title('Performance of Different Products in Terms of Sales')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Close the cursor and connection
cursor.close()
connection.close()
```

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_username', 'your_password', 'your_host', and 'your_database' with actual credentials
connection = mysql.connector.connect(
    user='root',
    password='ROOT',
    host='localhost',
    database='ecommerce'
)

# Create a cursor object to execute SQL queries
cursor = connection.cursor()

# Query to identify the most profitable products based on the difference between original and selling prices
most_profitable_products_query = '''
SELECT p.product_id, p.product_name, (SUM(od.total_price) - SUM(od.quantity * p.original_price)) as profit
FROM order_details od
JOIN product p ON od.product_id = p.product_id
GROUP BY p.product_id, p.product_name
ORDER BY profit DESC
LIMIT 10
'''

# Execute the query
cursor.execute(most_profitable_products_query)

# Fetch the result into a Pandas DataFrame
most_profitable_products_data = pd.DataFrame(cursor.fetchall(), columns=['Product ID', 'Product Name', 'Profit'])

# Plotting the most profitable products based on the difference between original and selling prices using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(most_profitable_products_data['Product Name'], most_profitable_products_data['Profit'], color='green')
plt.title('Most Profitable Products Based on the Difference between Original and Selling Prices')
plt.xlabel('Product Name')
plt.ylabel('Profit')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Close the cursor and connection
cursor.close()
connection.close()
```
